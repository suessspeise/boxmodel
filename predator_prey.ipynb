Prey:
$$
\dot{x}= ( b − py ) x
$$
Predator:
$$
\dot{y}= ( rx − d ) y
$$

The parameter $b$ is the growth rate of species $x$ (the prey) in the absence of interaction with species $y$ (the predators). Prey numbers are diminished by these interactions: The per capita growth rate decreases (here linearly) with increasing $y$ , possibly becoming negative.
 * The parameter $p$ measures the impact of predation on $\dot{x}/x$.
 * The parameter $d$ is the death (or emigration) rate of species $y$ in the absence of interaction with species $x$.
 * The term $rx$ denotes the net rate of growth (or immigration) of the predator population in response to the size of the prey population. 

```python
from boxmodel import BoxModel

b = 1.02
r = 1.05
d = 0.01


def prey(x,pred=0):
    return (b - pred) ** x

def predator(x, prey=0):
    return (r * prey - d) ** x

model = BoxModel(step=(500,0.1))

model.add_box('prey', {'n': 0.0})
model.add_box('predator', {'n': 0.0})
              #, 'c': material['c'], 'k': material['k']})

model.boxes['prey'].add_process('growth', 'n', prey, 
                                       [ model.ref('prey_n'),
                                         model.ref('predator_n')
                                       ])
model.boxes['predator'].add_process('growth', 'n', predator, 
                                       [ model.ref('predator_n'),
                                         model.ref('prey_n')
                                       ])

model.run()
model.plot()
```
